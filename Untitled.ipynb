```python
import nibabel as nib
from nilearn.image import load_img, resample_to_img, math_img, new_img_like
from nilearn.datasets import fetch_atlas_harvard_oxford

# Load the HCP NIfTI file (e.g., a functional or structural scan)
hcp_img = load_img('/scratch/alpine/alar6830/WM_All/100206/LR/tfMRI_WM_LR.nii.gz')

# Load the Harvard-Oxford atlas (1mm resolution)
atlas = fetch_atlas_harvard_oxford('cort-maxprob-thr50-1mm')
atlas_img = load_img(atlas['maps'])
atlas_labels = atlas['labels']

# Resample the atlas to match the HCP image resolution (1.6mm)
resampled_atlas = resample_to_img(atlas_img, hcp_img, interpolation='nearest')


```

```python
# Identify the indices corresponding to the PFC regions
pfc_labels = ['Frontal Pole', 'Superior Frontal Gyrus',
              'Middle Frontal Gyrus', 'Inferior Frontal Gyrus']
pfc_indices = [atlas_labels.index(label) for label in pfc_labels]

```

```python
# Create a binary mask for the PFC
pfc_mask_data = (nib.load(resampled_atlas).get_fdata() == pfc_indices[0]).astype(int)
for idx in pfc_indices[1:]:
    pfc_mask_data += (nib.load(resampled_atlas).get_fdata() == idx).astype(int)

pfc_mask_img = new_img_like(hcp_img, pfc_mask_data)

# Apply the PFC mask to the HCP data
pfc_hcp_img = math_img("img1 * img2", img1=hcp_img, img2=pfc_mask_img)

# Save the extracted PFC region
pfc_hcp_img.to_filename('hcp_pfc_extracted.nii.gz')

print("PFC extraction complete for HCP data!")
```
